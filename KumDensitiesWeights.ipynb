```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Kumaraswamy density function
def J(u, a, b):
    return a * b * u**(a - 1) * (1 - u**a)**(b - 1)

# Define the range of u
u = np.linspace(0.001, 0.999, 1000)

# These show the full flexibility: b<1, b>1, symmetric, U-shaped, peaked
params = [
    (2, 0.8),    # weights near u=1 (upper order statistics)
    (0.8, 2),    # weights near u=0 (lower order statistics)
    (4, 15),     # peaked/concentrated
    (1.2, 1.3),  # slightly peaked, near uniform
    (0.8, 0.8),  # U-shaped (both tails)
    (5, 5),      # symmetric bell-shaped
]

# Colors and line styles
colors = ['black', 'green', 'magenta', [0.4940, 0.1840, 0.5560], 'red', 'blue']
line_styles = ['-', '-', '-.', (0, (3, 1, 1, 1, 1, 1)), '--', ':']
line_widths = [1.5, 1.5, 1.5, 1.5, 1.5, 1.5]

# Create the plot
plt.figure(figsize=(6, 5))

# Plot J(u) for each pair of (a, b)
for (a, b), color, style, lw in zip(params, colors, line_styles, line_widths):
    plt.plot(u, J(u, a, b), label=fr'$(a, b) = ({a}, {b})$',
             color=color, linestyle=style, linewidth=lw)

# Add labels
plt.xlabel(r'$u$', fontsize=12)
plt.ylabel(r'$J(u; a, b)$', fontsize=12)
plt.ylim(0, 5)
plt.xlim(0, 1)

# Legend (two columns, upper area)
plt.legend(loc='upper left',
           bbox_to_anchor=(0.02, 0.98),
           fontsize=10,
           frameon=False,
           labelspacing=0.3,
           ncol=2,
           handlelength=2.5,
           columnspacing=1.0)

plt.tight_layout()

# Save as EPS and PNG
plt.savefig('KumDensities.eps', format='eps', dpi=300, bbox_inches='tight')
plt.savefig('KumDensities.png', format='png', dpi=300, bbox_inches='tight')

print("Saved: KumDensities.eps and KumDensities.png")
plt.show()
```
